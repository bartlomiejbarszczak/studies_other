```python
import copy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

# load the famous iris data
irisRaw = load_iris()

# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
display(iris)
```

### 3.

```python
display(iris.describe(include='all'))
```

### 4.

```python
new_iris = copy.deepcopy(iris)
new_iris = new_iris.groupby(['target']).mean()
display(new_iris)
```

```python
display(iris.head(n=5))
```

### 6.

```python
new_iris = iris.dropna(inplace=False)
display(new_iris)
if new_iris.shape == iris.shape:
    print("No missing values")
else:
    print("Missing values")
```

### 7.

```python
new_iris = iris.sort_values(by="sepal width (cm)", ascending=True, inplace=False)
display(new_iris)
```

### 8.

```python
min_index = iris[["petal length (cm)"]].idxmin()
max_index = iris[["petal length (cm)"]].idxmax()
print(f'Min index: {min_index[0]};\tmax index: {max_index[0]}')
```

### 9.

```python
new_iris = iris.std(axis=0)
display(new_iris)
```

### 10.

```python
new_iris = iris[(iris["sepal length (cm)"] > iris["sepal length (cm)"].mean())]
print(f'Mean value: {iris["sepal length (cm)"].mean()}')
display(new_iris)
```

### 11.

```python
target = "target"
n_bins = 10

for col_name in iris.columns:
    if col_name == target:
        continue

    target_values = list(set(iris[target]))
    fig, ax = plt.subplots(1, len(target_values), figsize=(15, 5))

    for index in range(0, len(target_values)):
        data = iris[(iris[target] == target_values[index])]
        ax[index].hist(data[col_name], bins=n_bins, edgecolor='black')
        ax[index].set_title(f'Target: {target_values[index]}')
        ax[index].set_xlabel("Length")
        ax[index].set_ylabel("Frequency")

    fig.suptitle(f"Histogram - {col_name}")
    plt.show()
```
